# 1. 팩터의 탄생 : CAPM & 3Factor Model

## 1.1. 시장 베타의 탄생, CAPM(자산가격결정모델)

- CAPM : 시장이 균형상태일 때, 자본자산의 균형가격이 어떻게 결정되는가에 관한 모형
- 균형 상태에서의 베타-기대수익률 사이의 관계 보여줌
- CML과 SML
    - 자본시장선 Capital Market Line, CML
        - 무위험 자산과 위험 자산의 조합으로 구성된 투자선
        - 투자자가 특정 위험 수준에서 예상 수익률을 최대화하는 포트폴리오
    - 증권시장선 Security Market Line, SML
        - 균형 자본시장에서 개별 증권의 체계적 위험(베타)와 기대수익률의 가격결정 관계를 설명해주는 선형 모델
        - 시장에서 거래되는 모든 자산에 대해 균형 상태에서의 위험과 수익률 간의 관계를 설명
 
    (출처 : 파이썬으로 배우는 포트폴리오)

```python
import pandas as pd
from scipy import stats, polyval, polyfit  # 통계도구, 다항식계산, fitting
from pylab import plot, title, show, legend # Numpy와 Matplotlib의 pyplot이라는 모듈을 같이 불러오기
import FinanceDataReader as fdr
import numpy as np
```

- 가격 데이터 불러오기
- 01. Market Data : Russell3000 
- 02. 개별 종목 Data : Apple 

```python
# 01. Russell3000(2010~현재) 수익률
russell = fdr.DataReader('VTHR', '2010')
russell = pd.DataFrame(russell['Close'])
russell_returns = russell['Close'].pct_change() # 주가 데이터 -> 수익률
russell_returns
```

```python
# 02. Apple(2010~현재) 수익률
apple = fdr.DataReader('AAPL', '2010')
apple = pd.DataFrame(apple['Close'])
apple_returns = apple['Close'].pct_change() # 주가 데이터 -> 수익률
apple_returns
```

```python
x = russell_returns.to_frame()
y = apple_returns.to_frame()

print(x)  # Russell3000 수익률
print('-----------------------')
print(y)  # Apple 수익률
```

```python
stocks = pd.concat([x, y], axis=1, join='inner')
stocks.columns = ['russell', 'apple']
```

```python
stocks = stocks.iloc[1:]
stocks
```

```python
x = stocks['russell'].values
y = stocks['apple'].values
```

- 기울기, y절편, r-value, p-value

* r-value == R-squared : 주식 가격의 변동을 설명하는 데 사용되는 회귀모델의 적합도 (0~1)
* p-value : 특정 가설의 유의도 검정 (< alpha 이면, 귀무가설(=0) reject, 해당 독립변수가 종속변수에 유의한 영향을 미친다고 판단)
* 추정오차 : 모델의 정확성 및 예측 신뢰성 평가 지표, 예측값과 실제값과 차이

```python
# 선형회귀모델
slope, intercept, r, p, stderr = stats.linregress(x, y)
ry = np.polyval([slope, intercept], x)   # 추정오차(잔차, residual)  # 다항식 : slope * x + intercept

# print(np.polyfit(x, y, 1))  # 기울기, y절편만 나오는 함수 
print("기울기 :", slope)
print("y절편 :", intercept)
print("r-value(R-squared) :", r)
print("p-value :", p)
print("표준 오차 :", stderr)
print("추정 오차:", ry)
```

## 1.2. APPLE - Russell3000 수익률 관계

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))

plot(x,y,'k.') #실제 데이터
plot(x,ry,'r-') #회귀선
title('Apple & Russell3000 regression (2010~2023)')
legend(['original','regression'])
show()
print(f'선형회귀분석 결과 : y = {intercept:.4f} + {slope:.4f}x')
```

### 1.2.1. APPLE - Russell3000 회귀분석 추정식

```python
# x : russell3000.values [r_i]
# y : apple.values [factor]

# 상관계수
corr = np.corrcoef(x,y)[1,0]   

# 표준편차
std_x = np.std(x)
std_y = np.std(y)

# 평균
mean_x = np.mean(x)
mean_y = np.mean(y)

# 베타
beta = corr*std_y / std_x

# 알파
alpha = mean_y - beta*mean_x

# 회귀분석 추정식
estimated_y = alpha + beta*x
print(f'회귀분석 추정식 : y = {alpha:.4f} + {beta:.4f}x')

# 회귀분석에서의 주요 관심사 : 독립변수-종속변수 간 관계 파악 
# -> 이때 회귀계수(기울기)가 0인지 여부를 알아야 함 
# 회귀계수 == 0 ; 독립변수-종속변수 간 관계 X
# 회귀계수 != 0 ; 독립변수-종속변수 간 유의미한 관계 O

# 상관계수의 t-test : 회귀계수(기울기)의 유의성 평가를 위해 진행
# t-test를 통해 상관계수의 유의성을 평가하여 p-value를 계산하고 
# 이를 통해 상관계수가 유의미한지 판단하는 데 사용
n = len(x)
t_statistic = corr * np.sqrt((n - 2) / (1 - corr**2))

# p-value
p = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=n-2))
print(f'p-value = {p:.4f}이므로 귀무가설 기각, 회귀계수(기울기)가 0이 아니라고 판단')
```

```python
plt.figure(figsize=(8,4))
title('Regression')
plot(x, y, 'k.')           # 실제 데이터 (검정색 점으로 표시)
plot(x,estimated_y,'r-')   # 회귀선 (빨간색 실선으로 표시)
legend(['original','regression'])
show()
```

### 1.2.2. APPLE - Russell3000 CAPM

```python
# data01 파마프렌치 3factor (출처 : Kenneth R.French-Data Library)
data01 = pd.read_csv('./F-F_Research_Data_Factors_daily(1).csv',index_col=0)
data01.index = pd.to_datetime(data01.index, format='%Y%m%d')
data01
# ri - rf, SMB, HML, rf
```

```python
stocks
```

```python
rf = data01['RF']
rf = rf.to_frame()
stocks = pd.concat([stocks, rf], axis=1, join='inner')
stocks
```

```python
z1 = stocks['russell'] - stocks['RF']
z2 = stocks['apple'] - stocks['RF']
print(z1)
print('---------------------')
print(z2)
```

```python
# 선형회귀모델 (CAPM)
# z1 == r_m - r_f
# z2 == r_i - r_f
slope, intercept, r, p, std = stats.linregress(z1,z2)
ry = np.polyval([slope, intercept], z1) # 잔차(추정오차)

print("기울기 :", slope)
print("y절편 :", intercept)
print("r-value(R-squared) :", r)
print("p-value :", p)
print("표준 오차 :", stderr)
print("추정 오차:", ry)
```

```python
# 기존 z1, z2 타입 확인
print(type(z1), type(z2))
print('--------------------')
# 회귀분석을 위한 데이터 타입 변경(series -> numpy array)
z1 = z1.values
z2 = z2.values
print(type(z1), type(z2))
```

```python
plt.figure(figsize=(8,4))
plt.plot(z1, z2,'k.') #실제 데이터
plt.plot(z1, ry,'r-') #회귀선

plt.title('apple & russell CAPM (2010~2023)')
plt.legend(['original','regression'])
plt.show()
print(f'CAPM : ri = {intercept:.4f} + {slope:.4f}(rm - rf)')
```

```python
# APPLE - Russell3000 수익률 회귀분석 결과 요약
regression = [intercept, slope, r*r]
df = pd.DataFrame([regression], index=list('값'),columns=['초과 수익률', 'B(MKT)', 'R-Squared'])
df
```

* SMB, HML | 포트폴리오 분석에 사용

* SMB Small Minus Big : 시가총액 기준으로 작은 규모의 주식 pfo와 큰 규모의 주식 pfo 간의 수익률 차이(보통 작은 규모 회사의 주식의 수익률은 큰 규모보다 높은 수익률 기대(작은 규모의 회사가 성장 가능성이 크고, 시장에 대한 정보가 제한적이므로))


* HML High Minus Low : 높은 가치의 주식 pfo와 낮은 가치의 주식 pfo 간의 수익률 차이(일반적으로, 높은 가치의 주식은 저평가 & 낮은 가치의 주식은 고평가되어 있을 가능성 높음)

* 소형주 (Small-Cap)  
시가총액 작은 기업 / 성장 가능성 높음 / 주식 시장에서 높은 수익률, 높은 변동성 가짐  
예 ) ZOOM  

* 대형주 (Large-Cap)  
시가총액 큰 기업, 이미 잘 알려진 대규모 기업 / 안정적이고 예측 가능한 성장 추구 / 주식 시장에서 안정성 높음  
예) Apple   
    
    
    * 기업규모 분류기준
    * 시가총액 상위 100위 : 대형주
    * 시가총액 상위 101~300위 : 중형주
    * 시가총액 상위 301위~ : 소형주


* 성장주 (Growth Stocks)
높은 성장 잠재력을 가진 기업 -> 매출&이익이 지속적으로 증가하는 기업  
주가가 기대 성장에 의해 높은 수준으로 평가됨    
'금리' 오르면 주가에 영향이 큼  
예) Tesla

* 가치주 (Value Stocks)  
저평가된 주식, 주가가 기업의 재무적 가치에 비해 낮게 평가되어 있는 기업  
주식 시장에서 소외된 기업으로 간주될 수 있음    
'금리'가 올라도 주가에 영향이 크지 않음  
예) Berkshire Hathaway Inc. (NYSE: BRK.A) - 워런 버핏이 이끄는 다양한 산업에 투자하는 투자 회사    
    
    
    * 일반적으로, Book to Market Ratio (Book Equtiy Market Ratio)를 반영하여 가치주 & 성장주를 구분

```python
# 누적 수익률 확인 (누적 수익률 그래프 시각화를 위한 과정)
# np.numprod() : 누적곱
# 백분율로 변환하여 누적곱 수행 : 누적수익률 계산
SMB = np.cumprod(data01['SMB']/100+1)
HML = np.cumprod(data01['HML']/100+1)
```

    * HML(빨간색) : 성장주 대비 가치주의 초과 수익률 (가치주 - 성장주 == High - Low)

## 1.4. APPLE - Russell3000 수익률 회귀분석 결과

* 파마-프렌치 3요인 모형 : CAPM의 확장형 모델, 주식 시장에서의 수익률을 설명하는 모델
* CAPM의 한계 보완
* 3가지 요소를 통해 기업의 수익률 변동성을 설명하고 pfo 관리에 활용 가능
    * 1. 시장 요인 Market Risk Facotr : rm - rf == 시장 수익률 - 무위험 수익률
        - 전체 주식 시장의 수익률 변동을 설명
        - 주식 시장 전반의 변동성과 시장 포트폴리으의 수익률에 따라 주식 수익률을 설명하는 데 사용
        - 주로 시장 pfo와의 상관관계를 통해 측정됨
    * 2. 소형주 - 대형주 요인 Size Risk Factor : SMB
        - '기업 시가총액 크기에 따른' 주식 수익률 차이를 설명
        - 일반적으로, 소형주는 시장에 비해 과소평과되어 높은 수익률을 보이는 경향 존재
        - 소형주 pfo와 대형주 pfo의 차이를 기반으로 측정됨
    * 3. 가치주 - 성장주 요인 Value Risk Factor : HML
        - '주식의 가격과 가치에 따른' 주식 수익률 차이를 설명
        - 가치주 ) 저평가된 주식, 시장에 비해 과소평가되어 높은 수익률을 보이는 경향
        - 성장주 ) 높은 시장 기대에 따라 고평가된 주식, 상대적으로 낮은 수익률을 보이는 경향
        - 가치주 pfo와 성장주 pfo 차이 기반으로 측정됨

* 데이터 : Apple, Russell3000 일일 주가 데이터 (2010~2023)

### 1.4.2. 3Factors

```python
import statsmodels.api as sm
```

```python
stocks
```

```python
data01
```

```python
Y = b
X = pd.concat([data01['Mkt'], data01[['SMB', 'HML']]], axis = 1, join = 'inner')
```

```python
# rm - rf
Y = pd.DataFrame(Y)
Y.columns
```

```python
Y['Rm - Rf'] = Y[0]  # 데이터프레임 Y의 컬럼명을 0 -> Rm-Rf 로 변경
Y.drop([0], axis = 1, inplace = True)
Y
```

```python
X
```

* sm_add_constant()함수 : 선형 회귀 모델링 구출 시 상수항(절편)을 포함하여 독립변수에 상수항을 추가하는 기능 수행
 * Y = beta0 + beta1(rm-rf) + beta2(SMB) + beta3(HML) + e
 * 위 3factor 모델에서 beta0을 추가

```python
X = sm.add_constant(X)  # 상수항 결합
X
```

```python
len(X), len(Y)
```

```python
# 인덱스 match를 위한 작업
temp = pd.concat([X, Y], axis=1, join='inner')
temp
```

```python
# 1990 이후 데이터만 사용
temp = temp['1990-01-01':]
temp
```

```python
# SMB, HML : 소수점으로 단위변환
temp['SMB'] = temp['SMB'] / 100
temp['HML'] = temp['HML'] / 100
temp
```

```python
X = temp[['const', 'Mkt', 'SMB', 'HML']]
Y = temp['Rm - Rf']
```

```python
X
```

```python
Y
```

```python
# 최소자승법을 통해 회귀 모델 생성
model = sm.OLS(Y,X)   
results = model.fit()
```

```python
print(results.summary())
```

```python
print('Alpha/Beta(M,S,H):')
print(results.params)
print('--------------------------------------')
print('R-squared:',results.rsquared)
print('----------------------------------------------------------------------------')
print(f'3Factors Model : r = {results.params[0]:.4f} + {results.params[1]:.4f}(rm - rf) + {results.params[2]:.4f}(SMB) + {results.params[3]:.4f}(HML) + e')
```

## 1.5. 결론

```python
# CAPM (CAPM은 시장요인만 존재)
R = np.array([intercept,slope,0,0,r_value**2])

# 3Factor
r = np.array([results.params[0],results.params[1],results.params[2],results.params[3],results.rsquared])

res = pd.DataFrame(data=[R,r], index =['CAPM','3Factor'], columns = ['Alpha','Mkt','SMB','HML','R-squared'])
res
```

* R-squared 값 : 3Factor > CAPM  
: 개별 주식 수익률에 대한 설명력은 3factor 모델이 더 좋다는 것을 알 수 있음
